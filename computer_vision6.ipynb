```python
sift_setup = {"detector": "sift", "confidence_threshold": 0.2}
orb_setup = {"detector": "orb", "confidence_threshold": 0.2}
orb_stitcher = stitching.Stitcher(**orb_setup)
sift_stitcher = stitching.Stitcher(**sift_setup)
result_sift = cv2.cvtColor(sift_stitcher.stitch(['/content/drive/MyDrive/19.jpg', '/content/drive/MyDrive/20.jpg']), cv2.COLOR_BGR2RGB)
plt.imshow(result_sift)
```

```python
result_orb = cv2.cvtColor(orb_stitcher.stitch(['/content/drive/MyDrive/19.jpg', '/content/drive/MyDrive/20.jpg']), cv2.COLOR_BGR2RGB)
plt.imshow(result_orb)
```
