```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import img_as_float
```

```python
def saturate_image(image_name, ratio=1.2):
  bgr_image = cv2.imread(image_name, 1)
  lab_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2LAB)
  l = np.array(lab_image[:, :, 0], np.float32)
  a = np.array(lab_image[:, :, 1], np.float32)
  b = np.array(lab_image[:, :, 2], np.float32)
  h, w = a.shape
  for i in range(h):
    for k in range(w):
      a[i, k] = min(255, a[i, k] * ratio)
  h, w = b.shape
  for i in range(h):
    for k in range(w):
      b[i, k] = min(255, b[i, k] * ratio)
  res = cv2.merge([l, a, b])
  res = cv2.cvtColor(res.astype('uint8'), cv2.COLOR_LAB2BGR)
  plt.imshow(res)
```

```python
def brightness_change(image_name, delta=50):
  image = cv2.imread(image_name, 1)
  lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
  l = np.array(lab_image[:, :, 0], np.float32)
  h, w = l.shape
  for i in range(h):
    for k in range(w):
      l[i, k] = min(255, l[i, k] + delta)
  a = np.array(lab_image[:, :, 1], np.float32)
  b = np.array(lab_image[:, :, 2], np.float32)
  res = cv2.merge([l, a, b])
  res = cv2.cvtColor(res.astype('uint8'), cv2.COLOR_LAB2BGR)
  plt.imshow(res)
```

```python
brightness_change('/content/drive/MyDrive/12.png', 150)
```
