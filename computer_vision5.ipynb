```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
clf = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascade_frontalface_default.xml')
image = cv2.cvtColor(cv2.imread('/content/drive/MyDrive/17.png', cv2.COLOR_BGR2RGB), cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
face = clf.detectMultiScale(gray_image, 1.1 , 9)
image_copy = image
for (x, y, w, h) in face:
    cv2.rectangle(image_copy, (x,y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(image_copy)
```
