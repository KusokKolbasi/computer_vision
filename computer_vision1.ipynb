```python
def get_average_brightness(img, left_bound, right_bound, bottom_bound, top_bound):
  if right_bound < left_bound or top_bound < bottom_bound:
    raise Exception('Wrong bounds!')
  gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  cropped_gray_image = gray_image[bottom_bound:top_bound, left_bound:right_bound]
  h, w = cropped_gray_image.shape
  denominator = h * w
  nominator = 0
  for i in range(h):
    for k in range(w):
      nominator += cropped_gray_image[i, k]
  return nominator / denominator
```

```python
left_bound = 100
right_bound = 700
bottom_bound = 100
top_bound = 700
ev = [np.arange(-5, 5+1)]
log_of_average_brightness = []
for i in range(11):
  log_of_average_brightness.append(math.log(get_average_brightness(images[i], left_bound, right_bound, bottom_bound, top_bound)))

print(ev)
print(log_of_average_brightness)
```

```python
plt.scatter(ev, log_of_average_brightness)
plt.xlabel('EV')
plt.ylabel('Log of average brightness')
plt.grid()
```
